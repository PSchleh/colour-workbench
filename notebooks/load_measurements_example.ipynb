# Example file read and analysis

This example notebook shows you how to read measurement files and do a few
useful things with them.

```python
from pathlib import Path

from specio.fileio import load_measurements

datapath = Path("Data")
datafile = datapath.joinpath("demo_measurements.csmf")
data = load_measurements(datafile)
print(data)
```

```python
first_measurement = data.measurements[0]
print(first_measurement)
```

## Example Data Usage
In this example we will extract the CIE 1931, 2-deg. chromaticity of all the 
measurements and then plot them. 

```python
from colour import XYZ_to_xy
from matplotlib import pyplot as plt
import numpy as np

XYZs = np.asarray([m.XYZ for m in data.measurements])
xy = XYZ_to_xy(XYZs)
fig = plt.figure()
plt.scatter(xy[:,0], xy[:,1])
ax = fig.gca()
ax.set_aspect(1)
```
